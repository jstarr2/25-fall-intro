```python
from datascience import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')

import warnings
warnings.simplefilter("ignore")
```

```python
Project_3_Data= Table.read_table('Project 3 data(Sheet1) (1).csv')
Project_3_Data.show()
```

```python
Data= Project_3_Data.select(0,8,11,14,17)
Data = Data.relabeled(0, 'ID')
Data = Data.relabeled(1, 'Enjoy Outdoors?')
Data = Data.relabeled(2, 'Cold or Warm?')
Data = Data.relabeled(3, 'Ranked Outdoor Activities')
Data = Data.relabeled(4, 'Hours outside a week')
Data.show()
```

```python
Grouped_Cold_or_warm= Data.group('Cold or Warm?')
Grouped_Cold_or_warm.show()
```

```python
Grouped_Cold_or_warm.barh('Cold or Warm?','count')
```

```python
Grouped_Enjoy= Data.group('Enjoy Outdoors?')
Grouped_Enjoy.show()
```

```python
Grouped_Enjoy.barh('Enjoy Outdoors?','count')
```

```python
Grouped_Hours= Data.group('Hours outside a week')
Grouped_Hours.show()
```

```python
Grouped_Hours.barh('Hours outside a week','count')
```
